```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress, ttest_ind
import numpy as np

# Bring in data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the data
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset with outerjoin on Mouse ID
combined_data_df = pd.merge(mouse_metadata, study_results,
                                 how='outer', on='Mouse ID')

# Display the data table for preview
combined_data_df.head()
```

```python
# Distinct count of mice
combined_data_df['Mouse ID'].nunique()
```

```python
# Count number of tests on the Mouse ID, top 5
combined_data_df['Mouse ID'].value_counts().head()
```

```python
#Check for any missing data
combined_data_df.isna().sum()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_data_df.loc[combined_data_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
print(duplicate_mice)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989 = combined_data_df.set_index('Mouse ID')
duplicate_mouse = g989.loc['g989', :]
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data_df = g989.drop(index = 'g989')
clean_study_data_df = clean_data_df.reset_index()
clean_study_data_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice = clean_study_data_df["Mouse ID"].nunique()
total_mice
```

## Summary Statistics

```python
# Show columns
print(clean_study_data_df.columns)
```

```python
# Calculate summary statistics by Drug Regimen
mean_tumor = clean_study_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_tumor = clean_study_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance_tumor = clean_study_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
stdev_tumor = clean_study_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem_tumor = clean_study_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Assemble the summary DataFrame
summary_df = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor,
    "Median Tumor Volume": median_tumor,
    "Tumor Volume Variance": variance_tumor,
    "Tumor Volume Std. Dev.": stdev_tumor,
    "Tumor Volume Std. Err.": sem_tumor
})

summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
summary_stats = clean_study_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
summary_stats
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.
# First create a dataframe grouping by the 'Drug Regimen' column
drug_group = clean_study_data_df.groupby('Drug Regimen')

# Create a dataframe counting the number of data points for each Drug Regimen
drug_group_count = pd.DataFrame(drug_group['Drug Regimen'].count())

# Create a bar chart based off of the group series from before and label the title
count_chart = drug_group_count.plot(kind='bar', title="Measurements Taken on Each Drug Regimen", color="blue")

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Drug Regimen")
count_chart.set_ylabel("Measurements")
plt.tight_layout()

# Show & save the figure
plt.savefig("output_data/Measurements Taken on Each Drug Regimen.png")
```
